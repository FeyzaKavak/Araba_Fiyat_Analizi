```python
# 1970 yila sahip araclari veri tabanindan cikarmayi deniyoruz

dataFrame = YuksekFiyatliAraclariSildik
dataFrame = dataFrame[dataFrame.year != 1970]
```

```python
dataFrame.groupby('year').mean()['price']
```

```python
#Transmission kolonu numeric olmadigi icin veri tabanindan cikarmayi deniyoruz
dataFrame = dataFrame.drop('transmission', axis=1)
#Veri Temizligini tamamladik
```

```python
# Model olusturuyoruz

y = dataFrame['price'].values
x = dataFrame.drop('price',axis=1).values
```

```python
from sklearn.model_selection import train_test_split # train ve test yapmak icin kütüphanemizi cagirdik
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequentialuential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12,activation='relu')) # 12 noron koyduk
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))

model.add(Dense(1)) #Cikis katmani

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x = x_train, y=y_train, validation_data=(x_test, y_test), batch_size=250, epochs=300)#epochs cok verilirse overfitting olabilir
#Egitim basliyor
```

```python
kayipVeri = pd.DataFrame(model.history.history)
```

```python
#loss ve validation loss u kıyaslıyoruz
kayipVeri.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahmin = model.predict(x_test)
```

```python
mean_absolute_error(y_test, tahmin) # 3356 pound luk bir sapma degeri soz konusu (yaklasik %13 luk bir fark)
```

```python
plt.scatter(y_test, tahmin)
plt.plot(y_test,y_test,'r')
```
